```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split, GridSearchCV
from sklearn.preprocessing import StandardScaler, OneHotEncoder
from sklearn.compose import ColumnTransformer
from sklearn.impute import SimpleImputer
from sklearn.pipeline import Pipeline
from sklearn.ensemble import RandomForestClassifier
from sklearn.linear_model import LinearRegression
from sklearn.metrics import classification_report, accuracy_score, confusion_matrix, mean_absolute_error, mean_squared_error
from sklearn.multioutput import MultiOutputRegressor
from sklearn.svm import SVC

# Load dataset
train_df = pd.read_csv('merged_features_drop_cleaned_train_data.csv')  # replace with your train dataset path
test_df = pd.read_csv('merged_features_drop_cleaned_test_data.csv')    # replace with your test dataset path

# Preprocessing function for the test data
def preprocess_data(df):
    # Convert datetime columns to datetime objects
    df['departure.scheduledTime'] = pd.to_datetime(df['departure.scheduledTime'], errors='coerce')
    df['arrival.scheduledTime'] = pd.to_datetime(df['arrival.scheduledTime'], errors='coerce')
    
    # Extract features from datetime columns
    df['departure.day_of_week'] = df['departure.scheduledTime'].dt.dayofweek
    df['departure.hour_of_day'] = df['departure.scheduledTime'].dt.hour
    
    # Fill missing values for weather-related columns
    weather_cols = ['Temperature (°F) Max', 'Temperature (°F) Avg', 'Temperature (°F) Min', 
                    'Dew Point (°F) Max', 'Dew Point (°F) Avg', 'Dew Point (°F) Min', 
                    'Humidity (%) Max', 'Humidity (%) Avg', 'Humidity (%) Min', 
                    'Wind Speed (mph) Max', 'Wind Speed (mph) Avg', 'Wind Speed (mph) Min', 
                    'Pressure (in) Max', 'Pressure (in) Avg', 'Pressure (in) Min', 
                    'Precipitation (in) Total']
    df[weather_cols] = df[weather_cols].fillna(df[weather_cols].mean())
    
    # Fill missing categorical columns (if any)
    categorical_cols = ['departure.iataCode', 'departure.icaoCode', 'airline.iataCode']
    df[categorical_cols] = df[categorical_cols].fillna('missing')

    # Drop unnecessary columns if they exist
    drop_columns = ['arrival.iataCode', 'arrival.icaoCode', 'arrival.estimatedTime']
    df.drop(columns=[col for col in drop_columns if col in df.columns], inplace=True)
    
    return df

# Preprocess the train and test data
train_df = preprocess_data(train_df)
test_df = preprocess_data(test_df)

# Define feature columns for model training
numeric_features = ['Temperature (°F) Max', 'Temperature (°F) Avg', 'Temperature (°F) Min', 
                    'Dew Point (°F) Max', 'Dew Point (°F) Avg', 'Dew Point (°F) Min', 
                    'Humidity (%) Max', 'Humidity (%) Avg', 'Humidity (%) Min', 
                    'Wind Speed (mph) Max', 'Wind Speed (mph) Avg', 'Wind Speed (mph) Min', 
                    'Pressure (in) Max', 'Pressure (in) Avg', 'Pressure (in) Min', 
                    'Precipitation (in) Total', 'departure.hour_of_day', 'departure.day_of_week']
categorical_features = ['departure.iataCode', 'departure.icaoCode', 'airline.iataCode']

# Define target for classification and regression tasks
train_df['binary_target'] = train_df['departure.delay_minutes'].apply(lambda x: 'delayed' if x > 0 else 'on-time')
train_df['multi_target'] = pd.cut(train_df['departure.delay_minutes'], 
                                   bins=[-np.inf, 0, 45, 175, np.inf], 
                                   labels=['No Delay', 'Short Delay', 'Moderate Delay', 'Long Delay'])
train_df['regression_target'] = train_df['departure.delay_minutes']

# Prepare features and targets for training
X_bin = train_df[numeric_features + categorical_features]
y_bin = train_df['binary_target']

X_multi = train_df[numeric_features + categorical_features]
y_multi = train_df['multi_target']

X_reg = train_df[numeric_features + categorical_features]
y_reg = train_df['regression_target']

# Build a function to create the model pipeline
def build_model_pipeline(model_type='binary'):
    # Preprocessing for both numeric and categorical data
    numeric_transformer = Pipeline(steps=[
        ('imputer', SimpleImputer(strategy='mean')),
        ('scaler', StandardScaler())
    ])
    
    categorical_transformer = Pipeline(steps=[
        ('imputer', SimpleImputer(strategy='constant', fill_value='missing')),
        ('onehot', OneHotEncoder(handle_unknown='ignore'))
    ])
    
    preprocessor = ColumnTransformer(
        transformers=[
            ('num', numeric_transformer, numeric_features),
            ('cat', categorical_transformer, categorical_features)
        ]
    )
    
    if model_type == 'binary':
        model = RandomForestClassifier()
    elif model_type == 'multi':
        model = RandomForestClassifier()
    elif model_type == 'regression':
        model = MultiOutputRegressor(LinearRegression())
    
    return Pipeline(steps=[('preprocessor', preprocessor), ('model', model)])

# Function to tune model hyperparameters
def tune_model(model, X, y):
    param_grid = {
        'model__n_estimators': [50, 100, 200],
        'model__max_depth': [None, 10, 20, 30],
        'model__min_samples_split': [2, 5, 10]
    }
    
    grid_search = GridSearchCV(model, param_grid, cv=5, n_jobs=-1, verbose=2)
    grid_search.fit(X, y)
    
    return grid_search.best_estimator_

# Train models for each task and make predictions
# 1. Binary Classification
binary_model = build_model_pipeline('binary')
best_binary_model = tune_model(binary_model, X_bin, y_bin)

# Make predictions for the test set (binary classification)
X_test_bin = test_df[numeric_features + categorical_features]
binary_preds = best_binary_model.predict(X_test_bin)

# Convert predictions to a string (on-time or delayed)
binary_preds = pd.Series(binary_preds, name='Status', dtype='str')

# Prepare submission file for binary classification
binary_submission = test_df[['flight.number', 'departure.iataCode', 'departure.icaoCode', 'departure.scheduledTime']]
binary_submission['Status'] = binary_preds
binary_submission.to_csv('binary_submission.csv', index=False)

# 2. Multi-Class Classification
multi_model = build_model_pipeline('multi')
best_multi_model = tune_model(multi_model, X_multi, y_multi)

# Make predictions for the test set (multi-class classification)
multi_preds = best_multi_model.predict(X_test_bin)

# Prepare submission file for multi-class classification
multi_preds = pd.DataFrame(multi_preds, columns=['Delay Category'])
multi_submission = test_df[['flight.number', 'departure.iataCode', 'departure.icaoCode', 'departure.scheduledTime']]
multi_submission['Delay Category'] = multi_preds['Delay Category']
multi_submission.to_csv('multi_class_submission.csv', index=False)

# 3. Regression Task
regression_model = build_model_pipeline('regression')
best_regression_model = tune_model(regression_model, X_reg, y_reg)

# Make predictions for the test set (regression)
regression_preds = best_regression_model.predict(X_test_bin)

# Prepare submission file for regression
regression_submission = test_df[['flight.number', 'departure.iataCode', 'departure.icaoCode', 'departure.scheduledTime']]
regression_submission['Delay'] = regression_preds[:, 0]  # Assuming only one output variable
regression_submission.to_csv('regression_submission.csv', index=False)

print("Submissions for Binary, Multi-Class, and Regression have been saved!")
```

```python
# --------------------------------------------
# Binary Classification for Flight Departure Delays
# Predicting "On-time" vs "Delayed"
# --------------------------------------------

# ----------------------------
# Phase 1: Import Libraries
# ----------------------------

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

from sklearn.model_selection import train_test_split, GridSearchCV, cross_val_score, KFold
from sklearn.preprocessing import StandardScaler, OneHotEncoder
from sklearn.compose import ColumnTransformer
from sklearn.pipeline import Pipeline
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score, classification_report, confusion_matrix
from sklearn.ensemble import RandomForestClassifier
from xgboost import XGBClassifier

import warnings
warnings.filterwarnings('ignore')  # Suppress warnings for cleaner output

# ----------------------------
# Phase 2: Load Data
# ----------------------------

# Load preprocessed training and testing datasets
train_df = pd.read_csv('merged_features_drop_cleaned_train_data.csv')
test_df = pd.read_csv('merged_features_drop_cleaned_test_data.csv')

print("Training Data Loaded:")
print(train_df.head())

print("\nTesting Data Loaded:")
print(test_df.head())

# ----------------------------
# Phase 3: Define Features and Target
# ----------------------------

# Define numerical and categorical features
numeric_features = ['Temperature (°F) Avg', 'Humidity (%) Avg', 'Wind Speed (mph) Avg', 'Pressure (in) Avg']
categorical_features = ['departure.iataCode', 'departure.icaoCode', 'airline.iataCode']

# Create binary target variable
# 0: on-time, 1: delayed
train_df['binary_target'] = train_df['departure.delay_minutes'].apply(lambda x: 0 if x == 0 else 1)

# Features (X) and Target (y)
X_bin = train_df[numeric_features + categorical_features]
y_bin = train_df['binary_target']

print(f"\nFeatures Shape: {X_bin.shape}")
print(f"Target Shape: {y_bin.shape}")

# ----------------------------
# Phase 4: Data Splitting
# ----------------------------

# Split data into training and validation sets (80% train, 20% validation)
X_train_bin, X_val_bin, y_train_bin, y_val_bin = train_test_split(
    X_bin, y_bin, test_size=0.2, random_state=42, stratify=y_bin
)

print(f"\nTraining Set Shape: {X_train_bin.shape}")
print(f"Validation Set Shape: {X_val_bin.shape}")

# ----------------------------
# Phase 5: Preprocessing Pipeline
# ----------------------------

# Define the preprocessing steps for numerical and categorical data
preprocessor = ColumnTransformer(
    transformers=[
        ('num', StandardScaler(), numeric_features),
        ('cat', OneHotEncoder(handle_unknown='ignore'), categorical_features)
    ]
)

# ----------------------------
# Phase 6: Model Training and Evaluation
# ----------------------------

# ----------------------------
# Step 6a: Random Forest Classifier
# ----------------------------

# Define the pipeline with Random Forest Classifier
rf_pipeline = Pipeline(steps=[
    ('preprocessor', preprocessor),
    ('classifier', RandomForestClassifier(random_state=42))
])

# Train the model
rf_pipeline.fit(X_train_bin, y_train_bin)

# Make predictions on the validation set
y_pred_rf = rf_pipeline.predict(X_val_bin)

# Evaluation Metrics for Random Forest
rf_accuracy = accuracy_score(y_val_bin, y_pred_rf)
rf_precision = precision_score(y_val_bin, y_pred_rf)
rf_recall = recall_score(y_val_bin, y_pred_rf)
rf_f1 = f1_score(y_val_bin, y_pred_rf)
rf_cm = confusion_matrix(y_val_bin, y_pred_rf)

print("\nRandom Forest Classifier Metrics:")
print(f"Accuracy: {rf_accuracy:.4f}")
print(f"Precision: {rf_precision:.4f}")
print(f"Recall: {rf_recall:.4f}")
print(f"F1-Score: {rf_f1:.4f}")
print("\nConfusion Matrix:")
print(rf_cm)
print("\nClassification Report:")
print(classification_report(y_val_bin, y_pred_rf))

# Plot Confusion Matrix
plt.figure(figsize=(6,4))
sns.heatmap(rf_cm, annot=True, fmt='d', cmap='Blues', xticklabels=['On-time', 'Delayed'], yticklabels=['On-time', 'Delayed'])
plt.title('Random Forest Confusion Matrix')
plt.xlabel('Predicted')
plt.ylabel('Actual')
plt.show()

# ----------------------------
# Step 6b: XGBoost Classifier
# ----------------------------

# Define the pipeline with XGBoost Classifier
xgb_pipeline = Pipeline(steps=[
    ('preprocessor', preprocessor),
    ('classifier', XGBClassifier(random_state=42, use_label_encoder=False, eval_metric='logloss'))
])

# Train the model
xgb_pipeline.fit(X_train_bin, y_train_bin)

# Make predictions on the validation set
y_pred_xgb = xgb_pipeline.predict(X_val_bin)

# Evaluation Metrics for XGBoost
xgb_accuracy = accuracy_score(y_val_bin, y_pred_xgb)
xgb_precision = precision_score(y_val_bin, y_pred_xgb)
xgb_recall = recall_score(y_val_bin, y_pred_xgb)
xgb_f1 = f1_score(y_val_bin, y_pred_xgb)
xgb_cm = confusion_matrix(y_val_bin, y_pred_xgb)

print("\nXGBoost Classifier Metrics:")
print(f"Accuracy: {xgb_accuracy:.4f}")
print(f"Precision: {xgb_precision:.4f}")
print(f"Recall: {xgb_recall:.4f}")
print(f"F1-Score: {xgb_f1:.4f}")
print("\nConfusion Matrix:")
print(xgb_cm)
print("\nClassification Report:")
print(classification_report(y_val_bin, y_pred_xgb))

# Plot Confusion Matrix
plt.figure(figsize=(6,4))
sns.heatmap(xgb_cm, annot=True, fmt='d', cmap='Oranges', xticklabels=['On-time', 'Delayed'], yticklabels=['On-time', 'Delayed'])
plt.title('XGBoost Confusion Matrix')
plt.xlabel('Predicted')
plt.ylabel('Actual')
plt.show()

# ----------------------------
# Phase 7: Hyperparameter Tuning
# ----------------------------

# ----------------------------
# Step 7a: Hyperparameter Tuning for Random Forest
# ----------------------------

# Define parameter grid for Random Forest
param_grid_rf = {
    'classifier__n_estimators': [100, 200, 300],
    'classifier__max_depth': [None, 10, 20],
    'classifier__min_samples_split': [2, 5, 10],
    'classifier__min_samples_leaf': [1, 2, 4]
}

# Initialize GridSearchCV for Random Forest
grid_search_rf = GridSearchCV(
    rf_pipeline,
    param_grid_rf,
    cv=5,
    scoring='f1',
    n_jobs=-1,
    verbose=2
)

# Fit GridSearchCV
print("\nStarting Grid Search for Random Forest...")
grid_search_rf.fit(X_train_bin, y_train_bin)

# Best parameters and score
print(f"\nBest Parameters for Random Forest: {grid_search_rf.best_params_}")
print(f"Best Cross-Validation F1 Score: {grid_search_rf.best_score_:.4f}")

# ----------------------------
# Step 7b: Hyperparameter Tuning for XGBoost
# ----------------------------

# Define parameter grid for XGBoost
param_grid_xgb = {
    'classifier__n_estimators': [100, 200, 300],
    'classifier__max_depth': [3, 5, 7],
    'classifier__learning_rate': [0.01, 0.1, 0.2],
    'classifier__subsample': [0.8, 1.0],
    'classifier__colsample_bytree': [0.8, 1.0]
}

# Initialize GridSearchCV for XGBoost
grid_search_xgb = GridSearchCV(
    xgb_pipeline,
    param_grid_xgb,
    cv=5,
    scoring='f1',
    n_jobs=-1,
    verbose=2
)

# Fit GridSearchCV
print("\nStarting Grid Search for XGBoost...")
grid_search_xgb.fit(X_train_bin, y_train_bin)

# Best parameters and score
print(f"\nBest Parameters for XGBoost: {grid_search_xgb.best_params_}")
print(f"Best Cross-Validation F1 Score: {grid_search_xgb.best_score_:.4f}")

# ----------------------------
# Phase 8: Final Model Selection and Evaluation
# ----------------------------

# Compare the best scores from both models
best_rf_score = grid_search_rf.best_score_
best_xgb_score = grid_search_xgb.best_score_

print(f"\nBest Random Forest F1 Score: {best_rf_score:.4f}")
print(f"Best XGBoost F1 Score: {best_xgb_score:.4f}")

# Select the best model based on GridSearchCV scores
if best_xgb_score > best_rf_score:
    best_model = grid_search_xgb.best_estimator_
    best_model_name = 'XGBoost (Tuned)'
else:
    best_model = grid_search_rf.best_estimator_
    best_model_name = 'Random Forest (Tuned)'

print(f"\nSelected Best Model: {best_model_name}")

# Make predictions on the validation set using the best model
y_pred_best = best_model.predict(X_val_bin)

# Evaluation Metrics for the Best Model
best_accuracy = accuracy_score(y_val_bin, y_pred_best)
best_precision = precision_score(y_val_bin, y_pred_best)
best_recall = recall_score(y_val_bin, y_pred_best)
best_f1 = f1_score(y_val_bin, y_pred_best)
best_cm = confusion_matrix(y_val_bin, y_pred_best)

print(f"\n{best_model_name} Metrics:")
print(f"Accuracy: {best_accuracy:.4f}")
print(f"Precision: {best_precision:.4f}")
print(f"Recall: {best_recall:.4f}")
print(f"F1-Score: {best_f1:.4f}")
print("\nConfusion Matrix:")
print(best_cm)
print("\nClassification Report:")
print(classification_report(y_val_bin, y_pred_best))

# Plot Confusion Matrix
plt.figure(figsize=(6,4))
if best_model_name.startswith('Random Forest'):
    cmap = 'Blues'
elif best_model_name.startswith('XGBoost'):
    cmap = 'Oranges'
else:
    cmap = 'Greens'

sns.heatmap(best_cm, annot=True, fmt='d', cmap=cmap, xticklabels=['On-time', 'Delayed'], yticklabels=['On-time', 'Delayed'])
plt.title(f'{best_model_name} Confusion Matrix')
plt.xlabel('Predicted')
plt.ylabel('Actual')
plt.show()

# ----------------------------
# Phase 9: Make Predictions on Test Data
# ----------------------------

# Ensure that test data has the same features as training data
# Reuse the preprocessor from the pipeline
# Select features
X_test_bin = test_df[numeric_features + categorical_features]

# Make predictions using the best model
y_test_pred_bin = best_model.predict(X_test_bin)

# Convert numeric predictions back to labels if necessary
# If the model outputs 0 and 1, map them to 'on-time' and 'delayed'
submission_binary = pd.DataFrame({
    'ID': range(1, len(y_test_pred_bin) + 1),
    'Delay': y_test_pred_bin
})

# Map numeric labels to string labels
submission_binary['Delay'] = submission_binary['Delay'].map({0: 'on-time', 1: 'delayed'})

# Preview the submission file
print("\nBinary Classification Submission File Preview:")
print(submission_binary.head())

# Save the submission file to CSV
submission_file_name_bin = 'submission_binary_classification.csv'
submission_binary.to_csv(submission_file_name_bin, index=False)
print(f"\nSubmission file '{submission_file_name_bin}' saved successfully.")

# ----------------------------
# Phase 10: Validate Submission File
# ----------------------------

# Check for missing values
print("\nMissing Values in Binary Classification Submission File:")
print(submission_binary.isnull().sum())

# Check unique values in 'Delay' column
print("\nUnique Values in 'Delay' Column:")
print(submission_binary['Delay'].unique())

# Verify the structure of the submission file
print("\nBinary Classification Submission File Info:")
print(submission_binary.info())
```
